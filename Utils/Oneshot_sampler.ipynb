```python
# Record 5s from mic and play it back

import sounddevice as sd
import soundfile as sf
import numpy as np
from IPython.display import display, Audio

SAMPLE_RATE = 22050

# Record audio
def record_audio(duration, samplerate=SAMPLE_RATE):
    audio = sd.rec(int(samplerate * duration), samplerate=samplerate, channels=2, dtype='float32')
    sd.wait()
    return audio

# Record for 5 seconds
duration = 1
print(f"Recording for {duration} seconds...")
audio_data = record_audio(duration)

# Play back the recorded audio in Jupyter

display(Audio(audio_data.T, rate=SAMPLE_RATE))
sf.write("testSamples\\noise.wav", audio_data, SAMPLE_RATE)
```
